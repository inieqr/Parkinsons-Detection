Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

Data Collection & Analysis

```python
# loading the data from csv file to a Pandas DataFrame
parkinsons_data = pd.read_csv('/content/parkinsons.csv')
```

```python
# printing the first 5 rows of the dataframe
parkinsons_data.head()
```

```python
# number of rows and columns in the dataframe
parkinsons_data.shape
```

```python
# getting more information about the dataset
parkinsons_data.info()
```

```python
# checking for missing values in each column
parkinsons_data.isnull().sum()
```

```python
# getting some statistical measures about the data
parkinsons_data.describe()
```

```python
# distribution of target Variable
parkinsons_data['status'].value_counts()
```

1  --> Parkinson's Positive

0 --> Healthy


```python
# grouping the data bas3ed on the target variable
parkinsons_data.groupby('status').mean()
```

Data Pre-Processing

Separating the features & Target

```python
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training

Support Vector Machine Model

```python
svm_model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
svm_model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
svm_train = svm_model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, svm_train)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on test data
svm_test = svm_model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, svm_test)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```

Classification Report

```python
print(classification_report(Y_test,  svm_test))
```

Confusion Matrix

```python
print(confusion_matrix(Y_test, svm_test))

df = {'y_Actual': Y_test, 'y_Predicted': svm_test}

df1 = pd.DataFrame(df, columns = ['y_Actual','y_Predicted'])

clf_confusion_matrix = pd.crosstab(df['y_Predicted'], df['y_Actual'], rownames = ['Predicted'], colnames=['Actual'])

sns.heatmap(clf_confusion_matrix, annot=True)
```

Logistic Regression Model

```python
lr_model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
lr_model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on training data
lr_train = lr_model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, lr_train)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on test data
lr_test = lr_model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, lr_test)
```

Classification Report

Confusion Matrix

```python
print(confusion_matrix(Y_test, lr_test))

df = {'y_Actual': Y_test, 'y_Predicted': lr_test}

df1 = pd.DataFrame(df, columns = ['y_Actual','y_Predicted'])

clf_confusion_matrix = pd.crosstab(df['y_Predicted'], df['y_Actual'], rownames = ['Predicted'], colnames=['Actual'])

sns.heatmap(clf_confusion_matrix, annot=True)
```

K-Nearest Neighbor Model



```python
knn_model = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
```

```python
# training the KNN model with Training data
knn_model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on test data
knn_test = knn_model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, knn_test)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```

Classification

Building a Predictive System

```python
input_data = (162.56800,198.34600,77.63000,0.00502,0.00003,0.00280,0.00253,0.00841,0.01791,0.16800,0.00793,0.01057,0.01799,0.02380,0.01170,25.67800,0.427785,0.723797,-6.635729,0.209866,1.957961,0.135242
)

# changing input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the data
std_data = scaler.transform(input_data_reshaped)

prediction = svm_model.predict(std_data)
print(prediction)


if (prediction[0] == 0):
  print("The Person does not have Parkinsons Disease")

else:
  print("The Person has Parkinsons")
```
